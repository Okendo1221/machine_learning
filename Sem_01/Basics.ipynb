### First, import some libraries

```
import numpy as np
```

```
N=10
array_1 = np.arange(N) # Create an array of N values
array_2 = np.random.randint(low=0,high=10,size=N) # Create an array of random int values
            # with the same lenght as array_1

print(array_1,array_2) # print both array_1 and array_2
print(array_1.shape, array_2.shape) # print both array_1 and array_2 shapes
print(np.sum(array_1),np.sum(array_2)) # print summation of array_1 and array_2
```

```
result_array = array_1*array_2 # Apply element-wise multiplication among array_1 and array_2
print("Result element-wise multiplication: ",result_array) # Print results
```

### Going further

```
# Messing everything around.
img = cv2.imread('/content/Onepiece.jpg')
# Convertir la imagen a escala de grises
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Mostrar la imagen en escala de grises
plt.imshow(img_gray, cmap='gray')
plt.axis('off')  # Opcional: para ocultar los ejes
plt.show()
# Ahora, aplica algunas tcnicas avanzadas:

# 1. Aplicar un filtro de suavizado (blur)
blurred_img = cv2.GaussianBlur(img_gray, (5, 5), 0)
plt.imshow(blurred_img, cmap='gray')
plt.title('Imagen Suavizada')
plt.axis('off')
plt.show()

# 2. Aplicar deteccin de bordes (Canny)
edges = cv2.Canny(img_gray, 100, 200)
plt.imshow(edges, cmap='gray')
plt.title('Deteccin de Bordes')
plt.axis('off')
plt.show()

# 3. Umbralizacin (Thresholding)
_, thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
plt.title('Umbralizacin')
plt.axis('off')
plt.show()

# 4. Rotar la imagen
(h, w) = img_gray.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_img = cv2.warpAffine(img_gray, M, (w, h))
plt.imshow(rotated_img, cmap='gray')
plt.title('Imagen Rotada')
plt.axis('off')
plt.show()

# 5. Histograma de intensidades
plt.hist(img_gray.ravel(), bins=256, range=[0,256])
plt.title('Histograma de Intensidades')
plt.show()
```

```
img_gray = cv2.imread('/content/Onepiece.jpg', cv2.IMREAD_GRAYSCALE)

# Modificar las intensidades de los pxeles
for r in range(img_gray.shape[0]):  # Recorrer filas
  for c in range(img_gray.shape[1]): # Recorrer columnas
    if img_gray[r, c] > 128: # Condicin simple: si el valor del pxel es mayor a 128
        img_gray[r, c] = 255 # Establecer el pxel a blanco
    else: # Do something
        img_gray[r, c] = 0

# Mostrar la imagen modificada
plt.imshow(img_gray, cmap='gray')
plt.axis('off')  # Opcional: para ocultar los ejes
plt.show()
```

### Save an image using OpenCV

### Crop the image - taking regions of interest (ROI)

```
# Extract some ROI
# Definir las coordenadas de la regin de inters (ROI) en la imagen original
x_start, y_start = 120, 80 # Coordenada superior izquierda (x, y)
x_end, y_end = 900, 600 # Coordenada inferior derecha (x, y)
# Extraer la ROI de la imagen en escala de grises
roi = img_gray[y_start:y_end, x_start:x_end]
plt.imshow(roi, cmap='gray')
plt.axis('off')  # Opcional: para ocultar los ejes
plt.show()
```

```
# Reemplazar una regin especfica en la imagen en escala de grises con la ROI
img_gray[y_start:y_end, x_start:x_end] = roi
# Mostrar la imagen en escala de grises con la regin reemplazada
plt.imshow(img_gray, cmap='gray')
plt.axis('off')  # Opcional: para ocultar los ejes
plt.show()
```
